```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
a = np.random.uniform(-10, 10, 500)
```

```python
# logarithmic
base = [50, 10, math.e, 0.1] #base can only be > 0 
a_nn = [i for i in a if i > 0] #log function only takes X > 0
plt.figure(figsize=(5, 6))
log_dict = {}
for i in base:
    log_dict[f'log{str(i)}_a'] = [*map(lambda x: math.log(x, i), a_nn)]
log_cmap = ['darkblue', 'lightblue', 'blue', 'steelblue',]
legends = []
for i, c in zip(log_dict.keys(), log_cmap):
    plt.scatter(a_nn, log_dict[i], color=c, alpha=0.8, linewidths=0.05)
    legends.append(mpatches.Patch(color=c, label=i))
    
log_dict[f'log_1+a'] = np.log1p([i for i in a if i > -1]) # 1+p moves x to x+1 so that even when x==0 this function is still valid 
plt.scatter([i for i in a if i > -1], log_dict['log_1+a'], color='salmon', alpha=0.8, linewidths=0.05)
legends.append(mpatches.Patch(color='salmon', label='log_1'))

plt.legend(handles=legends)
plt.title('logaithmic functions')
plt.show()
```

```python
# exponential
plt.figure(figsize=(5, 6))
base = [2, 1, 0, 0.5] # base can be >= 0
exp_dict = {}
for i in base:
    exp_dict[f'exp{str(i)}_a'] = [*map(lambda x: i**x, a)]
exp_cmap = ['darkblue', 'teal', 'lavender', 'lightskyblue', 'lavender']
legends = []
for i, c in zip(exp_dict.keys(), exp_cmap):
    plt.scatter(a, exp_dict[i], color=c, alpha=0.8, linewidths=0.05)
    legends.append(mpatches.Patch(color=c, label=i))
plt.legend(handles=legends)
plt.title('exponential functions')
plt.show()
```

```python
# polynomial

poly = [2, 1, 0, 0.5, 0.66, 3, 4, 5] # 'quardratic', 'linear', 1, 'squareroot', .
poly_dict = {}
for i in poly:
    poly_dict[f'poly{str(i)}_a'] = [*map(lambda x: x**i, a)]
poly_cmap = ['darkblue', 'teal', 'lavender', 'lightskyblue', 'royalblue']


fig, ax = plt.subplots(1,3,figsize=(15,6))
legends1 = []
for i, c in zip([2, 1, 0, 0.5, 0.66], poly_cmap):
    ax[0].scatter(a, poly_dict[f'poly{str(i)}_a'], color=c, alpha=0.8, linewidths=0.05)
    legends1.append(mpatches.Patch(color=c, label=f'poly{str(i)}_a'))
ax[0].legend(handles=legends1)

legends2 = []
for i, c in zip([0.5, 0.66], ['lightskyblue', 'royalblue']):
    ax[1].scatter(a, poly_dict[f'poly{str(i)}_a'], color=c, alpha=0.8, linewidths=0.05)
    legends2.append(mpatches.Patch(color=c, label=f'poly{str(i)}_a'))
ax[1].legend(handles=legends2)


legends3 = []
for i, c in zip([3, 4, 5], ['darkblue', 'teal', 'lavender']):
    ax[2].scatter(a, poly_dict[f'poly{str(i)}_a'], color=c, alpha=0.8, linewidths=0.05)
    legends3.append(mpatches.Patch(color=c, label=f'poly{str(i)}_a'))
ax[2].legend(handles=legends3)

fig.suptitle('polynomial functions')
plt.show()
```
